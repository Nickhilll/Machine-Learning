```python
# Replace missing values cat- mode, con -> mean and meadian
def replacer(df):
    cat = list(df.columns[df.dtypes == 'object'])
    
    for col in df.columns:
        
        if col in df.columns:
            mode = df[col].mode()[0]
            df[col] = df[col].fillna(mode)
        else:
            mean = df[col].mean()
            df[col] = df[col].fillna(mean)
```

## Apply One Hot Encoding on X_cat

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder(
    handle_unknown="ignore", sparse_output=False
).set_output(transform="pandas")

X_cat_pre = ohe.fit_transform(X_cat)
```

```python
X_cat_pre.head()
```

```python
df["Cylinders"].unique()
```

```python
X_con
```

## Standardizing is preffered

```python
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler().set_output(transform='pandas')
X_con_pre = scalar.fit_transform(X_con)
```

```python
X_con_pre.head()
```

## Combine the categorical and continuous variables   

```python
X_pre = X_con_pre.join(X_cat_pre)
X_pre.head()
```

## Step 5-> X_pre and Y
test_size = 0.2 to 0.3 

```python
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(X_pre, Y, test_size=0.2, random_state=21)
```

```python
from sklearn.metrics import (
    root_mean_squared_error,
    mean_absolute_error,
    mean_absolute_percentage_error,
    r2_score
)
```

```python
def evaluate_model(model, x, y):
    # predict results for x
    ypred = model.predict(x)
    
    # Calculate the metrics
    rmse = root_mean_squared_error(y,ypred)
    mae = mean_absolute_error(y, ypred)
    mape = mean_absolute_percentage_error(y, ypred)
    r2 = r2_score(y, ypred)
    
    # print above results
    print(f'RMSE : {rmse:.2f}')
    print(f'mae : {mae:.2f}')
    print(f'mape : {mape:.2%}')
    print(f'r2 : {r2:.2%}')
```

```python
print('Training Results:')
evaluate_model(model, xtrain, ytrain)
```
